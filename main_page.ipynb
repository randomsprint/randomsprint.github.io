# Random Sprint

Some processes are modeled as a "random walk," where a variables moves in a random direction independently of where it currently stands.

I have embraced the same phenomenon in my personal projects, but faster. I believe the best way to learn is to do, and here's a record of some of the doing I've done. The projects undertaken in my free time were completed quickly in order to learn new technologies, and I've often then implemented what I've learned in professional engagements.

Many of these projects are accessible through my [GitHub profile](https://github.com/randomsprint).

## Python Projects
### Valuations and "Beta Lag"

In this [workbook](http://htmlpreview.github.io/?https://github.com/moody-marlin/wmcm/blob/master/ddm_tests/Beta_Lag_Effect.html), I test one of the assumptions behind a widely used measure of industry risk in company valuations. I show that the assumption is not statistically rigorous in a way that likely biases company valuations in a downwards direction.

The analysis is supported by the [wmcm](https://github.com/moody-marlin/wmcm) package on github, which I developed with my friend [Chris White](https://github.com/moody-marlin).

### Shrimp Study

I analyzed the [GSMFC Seamap](http://www.gsmfc.org/seamap.php) database to look at changes in shrimp populations between various years. This was submitted to a federal court in relation to an economic damage claim.

### NFL Modeling Exercise

The goal of [this project](https://github.com/randomsprint/ddmnfl) is to create a Bayesian model of the NFL in [pymc](https://pymc-devs.github.io/pymc/) and use the posterior samples to test some theories I have. I haven't completed a write-up yet.

Currently relies on an old data source that only supplies final scores and ends at 2013. I'm going to create a better data source and then remodel with more data than just final scores.

An old write-up of my NFL modeling written in rjags can be found [here](http://www.dcfnerds.com/363/nfl-a-bayesian-model/).


### Analyzing Technical Indicators

In this project, I attempted to test "Technical Analysis" which is used to test stock prices. The results are hosted on my old blog [here](http://www.dcfnerds.com/analyzing-technical-analysis/).

This was my first python project. There is a lot I'd like to refactor.
