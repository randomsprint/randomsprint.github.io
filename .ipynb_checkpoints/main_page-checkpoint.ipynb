# Random Sprint

Some processes are modeled as a "random walk," where a variables moves in a random direction independently of where it currently stands.

I have embraced the same phenomenon in my personal projects, but faster. I believe the best way to learn is to do, and here's a record of some of the doing I've done.

## Python Projects
### Valuations and "Beta Lag"

In this [workbook](http://htmlpreview.github.io/?https://github.com/moody-marlin/wmcm/blob/master/ddm_tests/Beta_Lag_Effect.html), I test one of the assumptions behind a widely used measure of industry risk in company valuations. I show that the assumption is not statistically rigorous in a way that likely biases company valuations in a downwards direction.

The analysis is supported by the [wmcm](https://github.com/moody-marlin/wmcm) package on github, which I developed with my friend [Chris White](https://github.com/moody-marlin).

### NFL Correlation Study

When two teams play each other, you observe their relative strengths rather than their absolute strengths. If team A routs team B in a game expected to be tight, it could be that A is better than expected or it could be that B is worse than expected. The goal of this project is to create a Bayesian model of a league in [pymc](https://pymc-devs.github.io/pymc/) and use the posterior samples to predict teams which will over or under-perform together. This information would obviously be very useful to sports gamblers.

As a quick test of the idea, I applied the model seen in [this repository](https://github.com/mdekauwe/bayesian-football). The model is [specified here](https://github.com/randomsprint/ddmnfl/blob/master/exploratory/bf_recreation_2_refactor.ipynb) and [backtested here](https://github.com/randomsprint/ddmnfl/blob/master/exploratory/analysis.ipynb). My predictions didn't perform well in a backtest. I need to either try a different sport or develop a better model, which would require a better data source.
